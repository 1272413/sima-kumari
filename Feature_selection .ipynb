```python
df=pd.read_csv('paribas.csv',nrows=20000)
```

```python
df.shape
```

```python
x=df.drop('target',axis=1)
```

```python
x.shape
```

```python
x.info()
```

```python
cols=['int64','float64']
```

```python
numerical_columns=list(x.select_dtypes(include=cols).columns)
```

```python
len(numerical_columns)
```

```python
numerical_columns
```

```python
x=x[numerical_columns]
```

```python
x.shape
```

```python
x.info()
```

```python
correlated_features=set()
corelation_matrix=x.corr()
```

```python
corelation_matrix
```

```python
for i in range(len(corelation_matrix.columns)):
    for j in range(i):
        if abs(corelation_matrix.iloc[i,j])>0.8:
            colname= corelation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
len(correlated_features)
```

```python
correlated_features
```

```python
x=x.drop(correlated_features,axis=1)
```

```python
x.shape
```

```python
df=pd.read_csv('wisc_bc_data.csv')
```

```python
df.shape
```

```python
df
```

```python
x= df.drop(['id','diagnosis'],axis=1)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skf=SelectKBest(k=5,score_func=chi2)
```

```python
skf.fit(x,df['diagnosis'])
```

```python
skf.get_support()
```

```python
x.iloc[:,skf.get_support()]   # above 30 it select 
```
