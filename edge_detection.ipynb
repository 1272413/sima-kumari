```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('images/lenna.png',0)
```

```python
#sobel
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
combined_sobel= cv2.add(sobelx,sobely)
```

```python
final_img=np.hstack([sobelx,sobely,combined_sobel])
cv2.imshow('All image',final_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('original')
plt.imshow(img,cmap='gray')
plt.subplot(1,4,2)
plt.title('sobelx')
plt.imshow(sobelx,cmap='gray')
plt.subplot(1,4,3)
plt.title('sobely')
plt.imshow(sobely,cmap='gray')
plt.subplot(1,4,4)
plt.title('combine')
plt.imshow(combined_sobel,cmap='gray')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('original')
plt.imshow(img,cmap='gray')
plt.subplot(1,4,2)
plt.title('sobelx')
plt.imshow(sobelx,cmap='gray')
plt.subplot(1,4,3)
plt.title('sobely')
plt.imshow(abs(sobely),cmap='gray')
plt.subplot(1,4,4)
plt.title('combine')
plt.imshow(abs(combined_sobel),cmap='gray')
```

```python
import cv2
img=cv2.imread('images/lenna.png',0)
edges=cv2.Canny(img,50,150)
cv2.imshow('original',img)

cv2.imshow('canny Edge Detection',edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img2=cv2.imread('images/circle.png',0)
edges2=cv2.Canny(img2,50,150)
cv2.imshow('original',img2)

cv2.imshow('canny Edge Detection',edges2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('original')
plt.imshow(img2)
plt.subplot(1,4,2)
plt.imshow(edges2)
```
