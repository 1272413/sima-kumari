##### Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mtcars.csv')
```

```python
df
```

#####seperate the input and output var

```python
#disp, hp, wt
x=df[[
    'disp','hp','wt']] 
```

```python
x
```

```python
# mpg
y=df['mpg']
```

```python
y
```

```python
x.corrwith(y)
```

##### build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor= LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
regressor.score(x,y)
```

```python
regressor.coef_ ## all three are negativaly connected
```

```python
regressor.intercept_
```

##### Predict on unseen data

```python
disp= 221
hp=102
wt=2.91
```

```python
new= pd.DataFrame({
    'disp':[221],
    'hp':[102],
    'wt':[2.91]
})
new
```

```python
regressor.predict(new)
```

```python
disp*regressor.coef_[0]+hp*regressor.coef_[1]+wt*regressor.coef_[2]+regressor.intercept_
```
