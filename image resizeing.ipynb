```python
from skimage.transform import resize
import matplotlib.pyplot as plt
image=cv2.imread('images/lenna.png')
resized_image=resize(image,(1000,1400),anti_aliasing=True)
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)

plt.imshow(resized_image)
```

```python
#flipping
import cv2
image=cv2.imread('images/lenna.png')
flipped_horizontal=cv2.flip(image,0)
flipped_vertical=cv2.flip(image,1)
flipped_diagonal=cv2.flip(image,-1)
cv2.imshow('original',image)
cv2.imshow('afterhorizontal',flipped_horizontal)
cv2.imshow('afterver',flipped_vertical)

cv2.imshow('afterdia',flipped_diagonal)


cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
#flipping
flip_hor=cv2.cvtColor(flipped_horizontal,cv2.COLOR_BGR2RGB)
flip_ver=cv2.cvtColor(flipped_vertical,cv2.COLOR_BGR2RGB)
flip_dia=cv2.cvtColor(flipped_diagonal,cv2.COLOR_BGR2RGB)
img_org=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('horizontal')
plt.imshow(flip_hor)
plt.subplot(1,4,2)
plt.title('Ver')
plt.imshow(flip_ver)
plt.subplot(1,4,3)
plt.title('Dia')
plt.imshow(flip_dia)
plt.subplot(1,4,4)
plt.title('Original')
plt.imshow(img_org)
```
