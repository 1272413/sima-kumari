```python
#####frame scalling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

##### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape
```

```python
x_test.shape
```

##### Build the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

##### 
Predict on the unseen Data

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
y_test.value_counts()
```

##### Perfomance Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay 
```

```python
print(confusion_matrix(y_test,y_pred)) ## accuracy = 67+22=89% accuracy we get
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm.diagonal().sum()/cm.sum()
```

```python
print(classification_report(y_test,y_pred)) ##f1-Score is a harmonic mean formula= (2* precision *recall)/(precision +recall)
```

```python
##recall.0/ accuracy of calss 0/ sensitivity of algorithm
r0= 67/68
r0
```

```python
##recall.1/ accuracy of calss 1/ sensitivity of algorithm
r1= 22/32
r1
```

```python
## precision of class 0 /positive Predictive value
p0=67/77
p0
```

```python
## precision of class 1 /positive Predictive value
p1=22/23
p1
```

```python
# f1-score-0
2*p0*r0/(p0+r0)
```

```python
# f1-score-1
2*p1*r1/(p1+r1)
```
