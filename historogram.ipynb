```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num= [ np.random.randint(0,100) for x in range(1000)]
```

```python
print(num)
```

```python
plt.title('histogram')
plt.xlabel('value')
plt.ylabel('count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num,bins=range(10,101,10));
```
